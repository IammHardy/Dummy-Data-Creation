```python
import numpy as np
#create the variables
lowestnum =[-3, 0]
covariance_matrix=[[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
Rand_num= np.random.multivariate_normal(lowestnum,covariance_matrix,500)
print(Rand_num)
Rand_num.shape
```

# PROBLEM 2

```python

import matplotlib.pyplot as plt
%matplotlib inline

#plotting a graph to visualize the output of the doubling rate of chesnut buns with time.
plt.title(" visualization by scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(Rand_num[:,0],Rand_num[:,1])
```

```python
#VISUALIZE BY HISTOGRAM
import matplotlib.pyplot as plt
%matplotlib inline

#plotting a graph to visualize the output of the doubling rate of chesnut buns with time.
plt.title(" Histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(Rand_num[:,0])
plt.xlim(left=-5)
plt.show()
```

# PROBLEM 4

```python
import numpy as np
#create the variables
mean1 =[0, -3]
cov_m=[[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
randnum_1= np.random.multivariate_normal(mean1,cov_m,500)
print(randnum_1)
randnum_1.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

#plotting a graph to visualize the output of the doubling rate of chesnut buns with time.
plt.title(" Two in One Scatter Plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(Rand_num[:,0],Rand_num[:,1], s=20, label='0', marker='o')

plt.scatter(randnum_1[:,0], randnum_1[:,1], s=20, label='1', marker='o')
plt.legend()
plt.show()
```

# PROBLEM 5

```python
#DATA COMBINATION
data_combine = np.concatenate((Rand_num, randnum_1),axis=0)
print(data_combine)
data_combine.shape
```
